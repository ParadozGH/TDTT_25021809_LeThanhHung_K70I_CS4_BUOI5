```python
#W4A7
lst=[]
def checknt(n):
    check = True
    if n >= 0 and n < 2:
        check = False
    else:
        for i in range(2,int(n**0.5)+1):
            if n%i==0:
                check = False
                break
    return check
def uoc_prime(n):
    for i in range(1,int(n**0.5)+1):
        if n%i==0:
            lst.append(i)
            if n//i!=i:
                lst.append(n//i)
    maxx = 1
    for i in range(len(lst)):
        if lst[i] > maxx and checknt(lst[i]):
            maxx = lst[i]
    return maxx
n = int(input())
print(uoc_prime(n))
```

```python
#W4A8
def reversed_num(n):
    x = 0
    while n > 0:
        x = x*10 + n % 10
        n = n //10
    return x
def palindrome(i):
    if i == reversed_num(i):
        return True
    else:
        return False
n = int(input())
count = 0
while not palindrome(n):
    n += reversed_num(n)
    count += 1
print(count, n)
```

```python
#W4A11
def uoc_chan(n):
    cnt = []
    for i in range(1,int(n**0.5)+1):
        if n%i==0:
            cnt.append(i) if i%2==0 else None
            if n//i != i:
                cnt.append(n//i) if (n//i)%2==0 else None
    return len(cnt)
n = int(input())
print(uoc_chan(n))
```

```python
#W4A13
def tonguoc(n):
    sum = 0
    for i in range(1,int(n**0.5)+1):
        if n%i==0:
            sum+=i
            if n//i != i:
                sum+=n//i
    return sum - n
def check_relationship(a,b):
    if tonguoc(a) == b and tonguoc(b) == a:
        return True
    else:
        return False
a,b = map(int,input().split())
if check_relationship(a,b):
    print("true")
else:
    print("false")
```

```python
#W4A26
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
n = int(input())
for i in range(n):
    if fibonacci(i) > n:
        print(fibonacci(i-1), end=" ")
        break
```

```python
#W4A32
def password(s):
    cnt_db = 0
    cnt_thuong = 0
    cnt_hoa = 0
    cnt_so = 0
    check = True
    if len(s) <= 6:
        check = False
    else:
        for c in s:
            if c.isdigit():
                cnt_so += 1
            elif c.isalpha() and c.isupper():
                cnt_hoa += 1
            elif c.isalpha() and c.islower():
                cnt_thuong += 1
            elif not c.isalpha() and not c.isdigit():
                cnt_db += 1
        if cnt_hoa >= 1 and cnt_thuong >= 1 and cnt_so >= 1 and cnt_db >= 1:
            check = True
        else:
            check = False
    return check
s = input()
print("YES") if password(s) else print("NO")
```
