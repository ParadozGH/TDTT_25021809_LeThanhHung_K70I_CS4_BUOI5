```python
#W5A1
def Max_Of_Two(a,b):
    if a > b:
        return a
    else:
        return b
a,b = map(int,input().split())
print(Max_Of_Two(a,b))
```

```python
#W5A2
def swap(a,b):
    a = a^b
    b = b^a
    a = a^b
    print(a,b)
a,b = map(int,input().split())
swap(a,b)
```

```python
#W5A3
import math
def IsPrime(n):
    if n == 2:
        return True
    elif n < 2:
        return False
    else:
        for i in range(2,int(math.sqrt(n))+1):
            if n%i == 0:
                return False
        return True
n = int(input())
print(IsPrime(n))
```

```python
#W5A4
def IsPerfect(n):
    sum_divisors = 0
    for i in range(1,n):
        if n%i == 0:
            sum_divisors += i
    if sum_divisors == n:
        return True
    else:
        return False
n = int(input())
print(IsPerfect(n))
```

```python
#W5A5
n=list(map(int,input().split()))
k = int(input())
def FirstKValue(k):
    ans = -1
    for i in range(len(n)):
        if n[i] == k:
            ans = i
            break
    return ans
print(FirstKValue(k))
```

```python
#W5A6
memo = {}
def Factorial(n):
    if n in memo:
        return memo[n]
    if n == 0:
        memo[0] = 1
    else:
        memo[n] = n * Factorial(n - 1)
    return memo[n]
n = int(input())
print(Factorial(n))
```

```python
#W5A7
n = list(map(str,input().split()))
def Operation(n):
    if n[1] == '+':
        return f"{(float(n[0]) + float(n[2])):.2f}"
    elif n[1] =='-':
        return f"{(float(n[0]) - float(n[2])):.2f}"
    elif n[1] =='*':
        return f"{(float(n[0]) * float(n[2])):.2f}"
    elif n[1] =='/':
        return f"{(float(n[0]) / float(n[2])):.2f}"
print(Operation(n))
```

```python
#W5A8
def Denary_To_Binary(n):
    ans = ""
    while n > 0:
        ans = str(n % 2) + ans
        n //= 2
    return ans
def Hamming_Distance(x,y):
    if len(x) < len(y):
        x = '0' * (len(y) - len(x)) + x
    elif len(y) < len(x):
        y = '0' * (len(x) - len(y)) + y
    cnt = 0
    for i in range(len(x)):
        if x[i] != y[i]:
            cnt += 1
    return cnt
a, b = map(int,input().split())
c, d = Denary_To_Binary(a), Denary_To_Binary(b)
print(Hamming_Distance(c,d))
```

```python
#W5A9
def Sum_Of_Digits(n):
    ans = 0
    while n > 0:
        ans += n%10
        n = n//10
    return ans
n = int(input())
print(Sum_Of_Digits(n))
```

```python
#W5A10
def IsIsomorphic(s, t):
    if len(s) != len(t):
        return False
    map_st = {}  
    map_ts = {}  
    for c1, c2 in zip(s, t):
        if c1 in map_st:
            if map_st[c1] != c2:
                return False
        else:
            map_st[c1] = c2
        if c2 in map_ts:
            if map_ts[c2] != c1:
                return False
        else:
            map_ts[c2] = c1
    return True
a, b = map(str,input().split())
print(IsIsomorphic(a,b))
```
